### 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

### 2. Load Dataset

```python
df=pd.read_csv('placement.csv')
df
```

### 3. Explore Data

```python
df.shape
df.info()
df.describe()
```

```python
df.head()
```

### 4. Visualize Data

```python
sns.scatterplot(
    x=df['cgpa'],
    y=df['resume_score'],
    hue=df['placed'])
```

### 5. Split Features and Target

```python
''' Before fitting a Perceptron (or any supervised machine learning model),
we need to separate the dataset into:
Features (X) → Input data used to make predictions
Target (Y) → Output/label that we want to predict'''

X=df.iloc[:,0:2] #: → take all rows ,0:2 → take columns 0 and 1
Y=df.iloc[:,-1] # take the last column 
```

### 6. Train Perceptron

```python
from sklearn.linear_model import Perceptron 
p=Perceptron()
p.fit(X,Y)
```

### 7. Check Learned Parameters

```python
'''
So:
coef_ = slope information (weights)
intercept_ = bias/shift of the boundary
Together they define the decision boundary learned by the Perceptron. 
'''
print(p.coef_)  #returns the learned weights.
print(p.intercept_) #returns the bias term.
```

### 8. Training Accuracy

```python
print("Accuracy:", p.score(X, Y))
```

### 9. Visualize Decision Boundary

```python
from mlxtend.plotting import plot_decision_regions
#used to visualise the decision boundary learned by the Perceptron.

plot_decision_regions(X.values , Y.values, clf=p, legend=2) 
#clf- classifier, legend - controls how legend is to display
plt.xlabel("CGPA")
plt.ylabel("Resume Score")
plt.title("Perceptron Decision Boundary")
plt.show()
```
